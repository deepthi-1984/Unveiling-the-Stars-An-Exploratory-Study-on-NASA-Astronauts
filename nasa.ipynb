```python
import pandas as pd
```

```python
df=pd.read_csv('C:/Deepthi_Personal/DataAnalytics 2/Python/Nasa/nasa.csv')
df.head()
```

```python
null_values=df.isna().sum()
null_values
```

```python
df.dropna(subset=['Alma_Mater'],inplace=True)


filtered_rows=df.loc[df['Death_Mission'].isnull() & ~df['Death_Date'].isnull()]
df.drop(filtered_rows.index, inplace=True)

df['Birth_Date']=pd.to_datetime(df['Birth_Date'],dayfirst=True)
df['Death_Date']=pd.to_datetime(df['Death_Date'],dayfirst=True)
```

```python
df.to_csv('C:/Deepthi_Personal/DataAnalytics 2/Python/Nasa/astronaut.csv',index=False)
```
